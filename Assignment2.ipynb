# Task 1: Discount Rules (if / elif / else)
1. Write a program that reads an integer order_amount from the user using input().
2. Apply the following discount rules and print the final amount:
o order_amount >= 2000 15% discount
• 1500 <= order amount < 2000 → 10% discount o 1000 <= order amount < 1500 → 7% discount
otherwise 0% discount 
3. Ensure you convert input to a number and handle the case when the input is non-
numeric by displaying an error message and exiting the program.
Extra (optional): Add tax (fixed 5%) after discount and print subtotal, tax, and final total (still within conditionals and basic

```python
# 1. Write a program that reads an integer order_amount from the user using input().

order_amount = input("Enter Order Amount : ")

# 2. Apply the following discount rules and print the final amount:
# o order_amount >= 2000 15% discount
# • 1500 <= order amount < 2000 → 10% discount o 1000 <= order amount < 1500 → 7% discount
# otherwise 0% discount 

# 3. Ensure you convert input to a number and handle the case when the input is non-
# numeric by displaying an error message and exiting the program.
if order_amount.isnumeric():
    order_amount = int(order_amount)

    if order_amount >= 2000 :
        discount = 15
    elif order_amount >= 1500 :
        discount = 10
    elif order_amount >= 1000:
        discount = 7
    else :
        discount = 0
    
    discounted_amount = order_amount * (100 - discount)/100

    tax_amount = discounted_amount * 0.15

    final_amount = discounted_amount + tax_amount

    print(f"Discount amount on {order_amount } with {discount}% discount =  {discounted_amount} ")
    print(f"Tax of 15% = {tax_amount}")
    print(f"Final Amount = {final_amount}")
else : 
    print("Enter Numeric Value only !!! ")


# Extra (optional): Add tax (fixed 5%) after discount and print subtotal, tax, and final total (still within conditionals and basic
```

# Task 3: User Menu (while loop + break/continue)
1. Implement a simple while-loop driven menu that keeps asking the user to choose an
action until they press q to quit.

Menu options:

    1 - Add order amount to a running list (use input())
    2 - Show all orders and totals after applying discounts
    q -Quit
    
2. Use continue to re-show the menu after invalid input and break to exit on q
Constraints: Keep the loop logic simple and use only lists and loops (no functions required)

```python
#  Implement a simple while-loop driven menu that keeps asking the user to choose an
# action until they press q to quit.
# Menu options:
# 1 - Add order amount to a running list (use input())
# 2 - Show all orders and totals after applying discounts
# q -Quit

shopping_list_price = []

while True:
    option = input("""
                   1 - Add order amount \n
                   2 - Show all orders and totals after applying discounts \n
                   q -Quit \n
                   """)
    if option == "q":
        print("END")
        break
    elif option == "1":
        shopping_list_price.append(
            int(
                input("Enter Amount of Product : ")))
    elif option == "2":
        print(f"Shopping List : {shopping_list_price}")

        total_amount = 0
        discounted_amount = 0
        for order_amount in shopping_list_price:
            
            if order_amount >= 2000 :
                discount = 15
            elif order_amount >= 1500 :
                discount = 10
            elif order_amount >= 1000:
                discount = 7
            else :
                discount = 0
            total_amount += order_amount
            discounted_amount += order_amount *  discount /100
        print(f"total = {total_amount}")
        print(f"total discount = {discounted_amount}")

            
    else :
        print("Choose correct option ... ")
    
```

# Task 5: Loop Control with Conditions (break & continue)
1. Given a list of daily sales: daily = [200, 150, 0, 400, 50, -1, 300], iterate through it with a for loop and :
• If a day's value is -1, treat it as corrupted data and break the loop (stop processing).
o If a day's value is 0, treat it as a day with no sales and continue (skip adding to revenue )
• For valid positive sales, add to total _sales and print the running total.
2. Print the final total after the loop completes (or stops due to corrupted data)

```python
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0
for day in daily :
    if day == -1:
        total_sales = -1
        break
    elif day == 0 :
        continue
    else :
        total_sales += day
if total_sales == -1:
    print("corrupted data")
else :
    print(f"Total sales = {total_sales}")
```
