```python
#  From your products list, create a set of categories called categories_set. (If product
# names do not contain categories, create a short parallel list categories = [.] with
# matching length and use that.)
products = [
    ("milk", "Dairy"),
    ("bread","Food"),
    ("butter","Food"),
    ("tea powder","Beverages"),
    ("coffee","Beverages"),
    ("sugar","Food")
]

categories_set = set()

for product in products:
    categories_set.add(product[1])

print(categories_set)

categories_set.add("Food")

print(categories_set)


category = input("Enter a category name : ")

print(f"{category} is in {categories_set} =>  {category in  categories_set}")

# Extra (optional): Show how to get the total number of unique categories using a set.
print(f"Total categories : {len(categories_set)}")
```
