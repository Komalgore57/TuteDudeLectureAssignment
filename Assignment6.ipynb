# Task 2: 
Bill Calculator with Error Handling Given a list of product prices:
prices = [120, 350, 'abc', 500, -200, 800]
Write code that:
1. Iterates through the list.
2. Tries to add only valid (positive numerical) prices to the total. o TypeError if value is not a number
3. Handle
• Custom exception using raise ValueError("Negative price not allowed")
4. Prints the running total.
Exi ived belems but continue processing.

```python
# Bill Calculator with Error Handling Given a list of product prices:

prices = [120, 350, 'abc', 500, -200, 800]

# Write code that:
# 1. Iterates through the list.
total = 0
for price in prices:
    try : 
        i = int(price)
        if i<0:
            raise TypeError(f"{price} Negative Price Not allowed ")
    except ValueError :
        print(f"Price is not a Number {price}")
    except TypeError as e :
        print(e)
    else:
        total += i
print(f"Total : {total}")
```

# Task 3:
Custom Exception: Age Validator
1. Write a function check _age(age) that:

• Raises a custom exception ValueError("Age must be between 1 and 120") if age is out of range.

2. In your main code:
• Take age input from the user.
• Use try-except to catch and print the custom error message.

```python
def check_age(age):
    if age > 120 or age < 1 :
        raise ValueError("Age must be between 1 and 120")
    else:
        print(age)
        
age = int(input("Enter your age : "))
try : 
    check_age(age)
except ValueError as e:
    print(e)
```

# Task 4: 
File Reader with Exception Handling
1. Ask the user for a filename.
2. Try to open and read the file.
3. Handle:
    PermissionError
    File Not found Error
4. If successful, print first 3 lines of the file.
5. Use finally to print:
"File operation attempted."

```python
file_name = input("Enter file name : ")
try:
    file = open(file_name,"+r")
except PermissionError as pe:
    print(pe)
except FileNotFoundError as fnf:
    print(fnf)
else:
    data = file.readlines()[0:3]
    for i in data:
        print(i,end='')
```

Create a program that:

4. Inside the loop:
• Convert input to a float
• Handle ValueError if user enters invalid input

5. At the end, print:
o Total items o Total bill

```python
cart = []
item = 0
total = 0
while True:
    price = input("Enter price and 'q' to stop : ")
    if price =='q':
        break
    try:
        price = float(price)
        if price < 0:
            raise Exception("Invalid input")
    except ValueError:
        print("Price must be a number")
    except Exception as e:
        print(e)
    else:
        cart.append(price)
        total +=price
        item +=1
print(cart)
print(total)
print(item)

```
