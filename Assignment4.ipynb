# Task 1: 
Write Sales Records to a File
1. Create a list of sales amounts:
2. sales = [1200, 450, 980, 1500, 3000]
3. Open a file named sales_data.tt in write mode.
4. Write each sale on a new line in the file.
5. Close the file, then reopen it and print its contents.
Extra (optional): Write the data in comma-separated format instead of separate \n

```python
# 1. Create a list of sales amounts:
# 2. sales = [1200, 450, 980, 1500, 3000]

sales = [1200, 450, 980, 1500, 3000]

file = open("sales_data.txt","w")

for sale in sales:
    file.write(str(sale)+"\n")
file.close()

file = open("sales_data.txt","r")

data = file.readlines()

# data = data.split(",")

for i in data:
    print(i[:-1])

file.close()

#csv 

file = open("sales_data.csv","w")

for sale in sales:
    file.write(str(sale)+",")
file.close()
```

Task 4: Generate Summary Report from File
Using only file reading operations:

1. Read all sales values from sales_data.txt.

2. Convert them into integers.

3. Calculate and print:

      total sale
    • Highest Sale
    • Lowest Sale
    • Average Sale
    
Do not use any advanced libraries.

```python
file = open("sales_data.txt","r")
data = file.readlines()
print("\nData using readlines() in new list \n")
new_data = [ int(i[:-1]) for i in data]
print(new_data)
file.close()

total_sale = 0
highest_sale = new_data[0]
lowest_sale = new_data[0]
total_sale_count = 0

for sale in new_data:
    if sale < lowest_sale:
        lowest_sale = sale
    if sale > highest_sale:
        highest_sale = sale
    total_sale_count +=1
    total_sale += sale

print(f" highest sale : {highest_sale}")
print(f" lowest sale : {lowest_sale}")
print(f" total sale : {total_sale}")
print(f" avg sale : {total_sale / total_sale_count}")
```

# Task 5: 
Create Product Info File (User Input)
1. Ask the user for 3 product names & their prices.
2. Write them into a new file products.txt in this format:
product | price 
4. Read the file and print each line with proper format

```python
file_name = "products.txt"

file = open(file_name,"w")

for i in range(3):
    product = input("Enter Product name ")
    price = input("Enter product price ")
    file.write(product+ " | "+ price+"\n")
file.close()

# Read

file = open(file_name, "r")
print("Prodct Name. : Price")
data = file.readlines()
for i in data :
    product_price = i.split(" | ")
    print(f"{product_price[0]} :  {product_price[1][:-1]}")

file.close()
```

Task 7: Mini Project - Export Discounted Prices
Create a dictionary:

price = {
"Mouse": 500,
"Keyboard": 800,
"Monitor": 7000,
"Pendrive": 400,
"Camera": 5000
}
ASk the User for a discount %

write discounted prices into discount_report.txt using :
product | Orignal Price | Discounted Price

After writting read the and print on screen

Extra : 
Write Summary at the bottom of the the file

Total item : X
Average Discounted Price : Y



```python
prices = {
"Mouse": 500,
"Keyboard": 800,
"Monitor": 7000,
"Pendrive": 400,
"Camera": 5000
}

discount = int(input("Enter discount % : "))
file_name =  "discount_report.txt"

file = open(file_name, "w")
count = 0
total_discount_amount = 0
for price in prices.keys() :
    discount_amount = prices[price]*(100-discount)/100
    file.write(price+" | "+str(prices[price])+ " | "+str(discount_amount)+"\n")
    count += 1
    total_discount_amount += discount_amount

file.write(f"Total item : {count}\n")
file.write(f"Average Discounted Price : {total_discount_amount/count}")
file.close()
```
