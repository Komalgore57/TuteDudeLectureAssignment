#Task 1 - Basic Function: 

Price After Discount Write a function apply_discount(price, discount_percent) that:
1. Returns the price after discount.
2. If discount_percent is missing, apply a default discount of 5%.
3. Test the function with:
4. apply_discount(1000, 10)
5. apply_discount(500) # uses default discount
Extra (optional): Add a condition inside the function to ensure discount never exceeds 60%.

```python
def apply_discount(price, discount_percent=5):
    if discount_percent < 60:
        return price * (100 - discount_percent)/100
    else :
        return "NOT POSSIBLE"
        

print(apply_discount(100))
print(apply_discount(500,10))
print(apply_discount(500,100))


```

Recursive Function: Factorial Utility Create a recursive function factorial(n) that:
1. Returns the factorial of n.
2. Handles two edge cases: n == 0 and n == 1.
3. Prints an error message if n is negative.
Test with:
factorial(5)
factorial(0)
factorial(-3)

```python
def factorial(n):
    if n <0:
        return "N IS NEGATIVE"
    
    elif n == 0 or n ==1 :
        return 1
    
    else :
         return n * factorial(n-1)
    
print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]

prices_greater_than_500 = list(filter( lambda i : i > 500 , prices))
prices_less_than_and_eq_500 = list(filter( lambda i : i <= 500 , prices))

print(prices_greater_than_500)
print(prices_less_than_and_eq_500)
```

# Task 6 - Combined Utility Function
Write a function process_prices(prices) that:
1. Takes a list of prices.
2. Uses map() + lambda to apply 10% discount to all prices.
3. Uses filter() to keep only discounted prices that are above 300.
4. Returns both lists: o discounted_prices o filtered_prices
Test with:
process_prices([100, 500, 900, 50, 750])


```python
def process_prices(prices):
    # to apply 10% discount to all prices.
    discounted_prices = list (
        map(
            lambda x : x * .90 , prices
        )
    )
    # to keep only discounted prices that are above 300.
    filtered_prices = list(
        filter (
            lambda i : i >300 , prices
        )
    )

    # print(discounted_prices)
    # print(filtered_prices)
    return discounted_prices , filtered_prices

#. Test case

process_prices([100, 500, 900, 50, 750])
```

# Task 7 - Mini Problem: Menu Using Functions
Create the following three small functions:
1. add price(prices list, price) → adds price to the list.
2. get average_price(prices list) → returns average price.
3. get_max price(prices_list) → returns the maximum price.
Then create a simple menu:
1 → Add price
2 → Show average price
3 → Show highest price 
q → Quit
Use only loops + function calls (no OOP).


```python
# add price(prices list, price) → adds price to the list.
def add_price(prices_list, price):
    prices_list.append(price)
    return prices_list


# get average_price(prices list) → returns average price.

def get_average_price(prices_list):
    #return sum(prices_list) / len(prices_list)
    total_price = 0
    length =0
    for i in prices_list:
        total_price += i
        length += 1

    return total_price / length


# get_max price(prices_list) → returns the maximum price.

def get_max_price(prices_list):
    max_price = prices_list[0]
    for i in prices_list:
        if max_price < i :
            max_price = i
    return max_price

# Test cases
# print(add_price([1,2,3,4,5],6))
# print(get_average_price([10,10,10]))
# print(get_max_price([1,2,3,4,5]))

# Then create a simple menu:
# 1 → Add price
# 2 → Show average price
# 3 → Show highest price 
# q → Quit

list = []

while True:
    i = input ("""
1 → Add price
2 → Show average price
3 → Show highest price 
q → Quit
""")
    if i == "1" :
        value = int(input("Enter the number you want to add : "))
        add_price(list,value)
        print(list)

    elif i == "2":
        avg = get_average_price(list)
        print(avg)
    elif i == "3" :
        max_val = get_max_price(list)
        print(max_val)
    elif i =="q":
        break 
    
    else :
        continue
```
