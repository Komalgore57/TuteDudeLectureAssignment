Create a class Product with:

    name, price, category
Method  get_info() => print product details and ca o it detais
Extra (optional): Add a method apply_discount(percent) that returns the discounted price.

```python
class Product:
    def __init__(self,name,price, category):
        self.name = name
        self.price = price
        self.category = category
     
    def get_info(self):
  
        print(f"{self.name} | {self.price} | {self.category}")

    def apply_discount(self,percent):
        self.price = self.price * (100 - percent) /100

milk = Product("Milk", 35,"Dairy")
egg = Product("egg",250 , "Non-veg")

milk.get_info()
egg.get_info()
egg.apply_discount(10)
egg.get_info()
```
